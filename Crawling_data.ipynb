# **Pre-Processing**

```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
GazaHospital = data.drop(['tweet_url', 'conversation_id_str','user_id_str','lang','quote_count',
         'id_str', 'created_at'], axis=1)

GazaHospital
```

```python
!pip install emoji
```

```python
import re
import emoji

# Fungsi untuk membersihkan teks tweet
def cleanTweets(text):
    # Menghapus karakter selain huruf alfabet
    text = re.sub('[^a-zA-Z]', ' ', str(text).lower().strip())
    # Menghapus tag pengguna (@username)
    text = re.sub('@[A-Za-z0-9_]+', '', text)
    # Menghapus tanda pagar (#)
    text = re.sub('#', '', text)
    # Menghapus retweet (RT)
    text = re.sub('RT[\s]+', '', text)
    # Menghapus URL
    text = re.sub('https?:\/\/\S+', '', text)
    # Mengganti emoji dengan teks kosong
    text = emoji.demojize(text)
    # Mengganti semua emoji (termasuk yang masih berformat emoji) dengan teks kosong
    text = re.sub(r'(:[^:]*:)', '', text)
    # Mengganti karakter baris baru dengan spasi
    text = re.sub('\n', ' ', text)
    return text

# Menerapkan fungsi membersihkan teks ke kolom 'full_text' dan menyimpan hasilnya di kolom 'cleanedTweet'
GazaHospital['cleanedTweet'] = GazaHospital['full_text'].apply(cleanTweets)
```

```python
import re

# Fungsi untuk menghapus tanda baca
def remove_punctuation(text):
    # Menggunakan ekspresi reguler untuk menghapus tanda baca
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    return cleaned_text

# Membersihkan tanda baca pada kolom teks
GazaHospital['full_text']  = GazaHospital['full_text'].apply(lambda x: remove_punctuation(x))
```

```python
# Menampilkan DataFrame dengan teks tweet yang telah dibersihkan
GazaHospital
```

```python
import nltk
nltk.download('punkt')
```

```python
#tokenisasi untuk memisahkan kata
GazaHospital['text_tokens'] = GazaHospital['full_text'].apply(lambda x: word_tokenize(x))
```

```python
GazaHospital
```

```python
import nltk
nltk.download('stopwords')
```

```python
# Stopword removal

stop_words = set(stopwords.words('english'))

GazaHospital['text_tokens'] = GazaHospital['text_tokens'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
GazaHospital
```

```python
# Stemming
porter = PorterStemmer()
GazaHospital['stemmed_tokens'] = GazaHospital['text_tokens'].apply(lambda x: [porter.stem(word) for word in x])
```

```python
GazaHospital
```

```python
from wordcloud import WordCloud

# Menggabungkan semua token menjadi satu string
all_text = ' '.join([' '.join(tokens) for tokens in GazaHospital['stemmed_tokens']])

# Membuat word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Menampilkan word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from collections import Counter

# Menggabungkan semua token menjadi satu list
all_tokens = [token for sublist in GazaHospital['stemmed_tokens'].tolist() for token in sublist]

# Menghitung frekuensi kata-kata
word_frequencies = Counter(all_tokens)

# Mengambil kata dengan frekuensi tertinggi
top_words = dict(sorted(word_frequencies.items(), key=lambda x: x[1], reverse=True)[:10])

# Plotting bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_words.keys(), top_words.values(), color='skyblue')
plt.xlabel('Kata')
plt.ylabel('Frekuensi')
plt.title('Kata yang Paling Sering Muncul')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def label_sentiment(tokens):
    text = ' '.join(tokens)  # Menggabungkan daftar kata-kata menjadi sebuah string
    sentiment_scores = analyzer.polarity_scores(text)
    compound_score = sentiment_scores['compound']

    if compound_score >= 0.05:
        return "Positif"
    elif compound_score <= -0.05:
        return "Negatif"
    else:
        return "Netral"

GazaHospital['sentiment_label'] = GazaHospital['stemmed_tokens'].apply(label_sentiment)
```
